```python
job_cat = pd.DataFrame(cleandata.groupby('position').count()).reset_index()
job_cat.head(10)
```

```python
plt.figure(figsize=(13,8))
fig,ax = plt.subplots()

ax.barh(job_cat.position, job_cat.location, color='b', alpha=0.5, align="center")

plt.title("Job Openings for Job title")
ax.set_xlabel("Number of Positions ")

for i, v in enumerate(job_cat.position):
    ax.text(v + 30, i, str(v), color='black', fontweight='bold')
plt.grid()
plt.savefig("Bar_JobCounts.png", bbox_inches='tight')
# plt.tight_layout()
plt.show()
```

```python
regimen_datapoints = job_cat[["position"]]
regimen_datapoints = job_cat.rename(columns={"description": "count"})
regimen_datapoints = regimen_datapoints.set_index("position")
regimen_datapoints.plot(kind="bar", figsize=(10,6))

plt.title("Count per Job Title")

plt.show()
plt.tight_layout()
plt.savefig("2018 Job Count Bar Chart ")
```

```python
plt.figure(figsize=(10,6))
ax1 = plt.subplot(121, aspect='equal')
job_cat.plot(kind='pie', y = "description", ax=ax1, autopct='%1.1f%%', 
 startangle=90, shadow=False, labels=job_cat['position'], legend = True, )
plt.legend(bbox_to_anchor=(1,1.025), loc="lower left")
plt.savefig("2018 Job Count Pie Chart")
```
