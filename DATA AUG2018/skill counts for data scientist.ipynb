```python
import matplotlib.pyplot as plt
import pandas as pd
import requests
import numpy as np
import time
import os
import seaborn as sb
```

```python
DS = pd.read_csv("Data Scientist.csv")
DS
```

```python
DS['SS_Python_flag'] = 0 
DS['SS_SQL_flag'] = 0 
DS['SS_JavaScript_flag'] = 0 
DS['SS_MongoDB_flag'] = 0 
DS['SS_Hadoop_flag'] = 0 
DS['SS_Database_flag'] = 0 
DS['SS_ExcelVBA_flag'] = 0 
DS['SS_R_flag'] = 0 
DS['SS_Spark_flag'] = 0 
DS['SS_Tableau_flag'] = 0 
DS['SS_D3_flag'] = 0 
DS.head()

```

```python
for irow in np.arange(len(DS)):
    

    if (str("python") in DS.description[irow]):
        DS.iloc[irow,5] = 1
    
    if (str("sql") in DS.description[irow]) :
        DS.iloc[irow,6] = 1
        
    if (str("java") in DS.description[irow]):
        DS.iloc[irow,7] = 1
        
    if (str("mangodb") in DS.description[irow]) :
        DS.iloc[irow,8] = 1
    
    if (str("hadoop") in DS.description[irow]):
        DS.iloc[irow,9] = 1  
        
    if (str("database") in DS.description[irow]):
        DS.iloc[irow,10] = 1
    
    if ((str("excel") in DS.description[irow]) or ("vba" in DS.description[irow])) :
        DS.iloc[irow,11] = 1
             
    if (str("spark") in DS.description[irow]) :
        DS.iloc[irow,12] = 1
    
    if (str("tableau") in DS.description[irow]):
        DS.iloc[irow,13] = 1  
        
    if (str("d3") in DS.description[irow]):
        DS.iloc[irow,14] = 1  
    
TT_keywords = ['Python', 'SQL', 'Java', 'MangoDB','Hadoop','Database','Excel/VBA','R','Spark','Tableau','D3']
TT_Count = np.arange(len(TT_keywords))

for i_tech_tool in np.arange(len(TT_keywords)):
    TT_Count[i_tech_tool] = DS.iloc[:,5+i_tech_tool].sum()

print(TT_Count)

# construct a dataFrame using the TT keywords and TT keyword counSS
df_TT = pd.DataFrame({'TT_Keywords':TT_keywords, 'TTkeywords_Counts':TT_Count})
df_TT.sort_values('TTkeywords_Counts',inplace=True)
```

```python
plt.figure(figsize=(13,8))
fig,ax = plt.subplots()

ax.barh(df_TT.TT_Keywords, df_TT.TTkeywords_Counts, color='b', alpha=0.5, align="center")

plt.title("Popularity of Desired Technical Tools")
ax.set_xlabel("Number of Job Openings Requiring the Technical Tools")

for i, v in enumerate(df_TT.TTkeywords_Counts):
    ax.text(v + 30, i, str(v), color='black', fontweight='bold')
plt.grid()
plt.savefig("Bar_Popularity_of_Desired_Technial_Tools_2018.png", bbox_inches='tight')
# plt.tight_layout()
plt.show()
```
